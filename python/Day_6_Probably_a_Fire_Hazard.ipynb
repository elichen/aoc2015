```python
data = """turn on 0,0 through 999,999
toggle 0,0 through 999,0
turn off 499,499 through 500,500""".split('\n')
```

```python
data = open("input.txt").read().splitlines()
```

```python
import numpy as np

m = np.zeros((1000,1000))
for line in data:
  start, _, end = line.split()[-3:]
  sx,sy = list(map(int,start.split(',')))
  ex,ey = list(map(int,end.split(',')))
  if line.startswith("turn on"):
    m[sy:ey+1,sx:ex+1] = 1
  elif line.startswith("turn off"):
    m[sy:ey+1,sx:ex+1] = 0
  else:
    m[sy:ey+1,sx:ex+1] = np.logical_not(m[sy:ey+1, sx:ex+1]).astype(int)

import matplotlib.pyplot as plt

plt.imshow(m, cmap='gray')
plt.colorbar()
plt.show()

np.sum(m == 1)
```

```python
m = np.zeros((1000,1000))
for line in data:
  start, _, end = line.split()[-3:]
  sx,sy = list(map(int,start.split(',')))
  ex,ey = list(map(int,end.split(',')))
  if line.startswith("turn on"):
    m[sy:ey+1,sx:ex+1] = m[sy:ey+1,sx:ex+1] + 1
  elif line.startswith("turn off"):
    m[sy:ey+1,sx:ex+1] = np.maximum(m[sy:ey+1,sx:ex+1] - 1, 0)
  else:
    m[sy:ey+1,sx:ex+1] = m[sy:ey+1,sx:ex+1] + 2

int(np.sum(m))
```
