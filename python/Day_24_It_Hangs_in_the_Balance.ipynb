```python
data = """1
2
3
5
7
13
17
19
23
29
31
37
41
43
53
59
61
67
71
73
79
83
89
97
101
103
107
109
113""".split('\n')
data = set(map(int,data))
```

```python
import itertools
from functools import reduce
from tqdm.notebook import tqdm

def find(data):
  for i in tqdm(range(1,len(data))):
    for s1 in itertools.combinations(data, i):
      remain =  data-set(s1)
      if sum(s1)*2 != sum(remain): continue
      smallest_q = float("inf")
      for j in range(1,len(remain)//2):
        a = itertools.combinations(remain, j)
        for s2 in a:
          s3 = remain-set(s2)
          if sum(s1) == sum(s2) == sum(s3):
            # print(s1,s2,s3)
            smallest_q = min(reduce(lambda x,y: x*y, s1), smallest_q)
      if smallest_q != float("inf"):
        return smallest_q
find(data)
```

```python
def find2(data):
  for i in tqdm(range(1,len(data))):
    for s1 in itertools.combinations(data, i):
      remain =  data-set(s1)
      sum1 = sum(s1)
      if sum1*3 != sum(remain): continue
      smallest_q = float("inf")
      for j in range(1,len(remain)//2):
        for s2 in itertools.combinations(remain, j):
          sum2 = sum(s2)
          if sum2 != sum1 or sum1*2 != sum(remain-set(s2)): continue
          for k in range(1,len(remain-set(s2))//2):
            for s3 in itertools.combinations(remain-set(s2),k):
              sum3 = sum(s3)
              if sum3 != sum1: continue
              s4 = remain-set(s2)-set(s3)
              # print(s1,s2,s3,s4)
              if sum1 == sum2 == sum3 == sum(s4):
                smallest_q = min(reduce(lambda x,y: x*y, s1), smallest_q)
      if smallest_q != float("inf"):
        return smallest_q
find2(data)
```
