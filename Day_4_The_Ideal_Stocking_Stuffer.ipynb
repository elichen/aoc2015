```python
import hashlib
from tqdm.notebook import tqdm

def md5(string):
    return hashlib.md5(string.encode()).hexdigest()

key = "iwrupvqb"
for i in tqdm(range(10**9)):
  s = f"{key}{i}"
  if md5(s).startswith("00000"):
    print(i)
    break
```
